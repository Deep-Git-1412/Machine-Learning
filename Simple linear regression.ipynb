# simple linear regression single variable







```python
# Step:1 first we importing the useful libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#step:2 import or create the data

x_area=[2600,3000,3200,3600,4000,4100,4400,4700,5000]#independent variable
y_price=[550000,565000,610000,680000,725000,800000,830000,860000,900000]#dependent variable
data=list(zip(x_area,y_price))
df=pd.DataFrame(data,columns=['area','price'])
print(df)
```

```python
#step:3 visulize the data

plt.xlabel('AREA')  #label for x
plt.ylabel('PRICE') #label for y
plt.scatter(df.area,df.price,color='red') #scatter plotting
plt.show()
```

```python
#step 4:creating a object of linear regression

reg=linear_model.LinearRegression()
```

```python
#step 5:fitting our data into the model

reg.fit(df[['area']],df.price)
```

```python
#step 6:make a prediction on unseen or new data

reg.predict([[3300]])
```

```python
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.scatter(df.area,df.price,color='red')
plt.plot(df.area,reg.predict(df[['area']]))  #regression line
plt.show()
```
